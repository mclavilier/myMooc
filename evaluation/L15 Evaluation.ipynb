### Precision definition : 
Compute the precision

The precision is the ratio tp / (tp + fp) where tp is the number of true positives and fp the number of false positives. <B>The precision is intuitively the ability of the classifier not to label as positive a sample that is negative.
</B>

The best value is 1 and the worst value is 0.

### Recall Definition :
Compute the recall

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. <B>The recall is intuitively the ability of the classifier to find all the positive samples.
</B>

The best value is 1 and the worst value is 0.

### F1 Score Definition:
Compute the F1 score, also known as balanced F-score or F-measure

The F1 score can be interpreted as a weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal. The formula for the F1 score is:

F1 = 2 * (precision * recall) / (precision + recall)

In the multi-class and multi-label case, this is the average of the F1 score of each class with weighting depending on the average parameter.
